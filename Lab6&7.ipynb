```python
#Q1
import numpy as np
import scipy.stats as stats

sample = np.array([3.3, -0.3, -0.6, -0.9])
n = len(sample)
mean_sample = np.mean(sample)
sigma = 3
confidence_level = 0.90

# When sigma is known (use z-distribution)
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error_known = z_critical * (sigma / np.sqrt(n))
ci_known = (mean_sample - margin_of_error_known, mean_sample + margin_of_error_known)

# When sigma is unknown (use t-distribution)
sample_std = np.std(sample, ddof=1)
t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=n-1)
margin_of_error_unknown = t_critical * (sample_std / np.sqrt(n))
ci_unknown = (mean_sample - margin_of_error_unknown, mean_sample + margin_of_error_unknown)

print(f"90% Confidence Interval (σ known): {ci_known}")
print(f"90% Confidence Interval (σ unknown): {ci_unknown}")
```

```python
#5
import numpy as np
import scipy.stats as stats

journals = np.array([1.79, 1.75, 1.67, 1.65, 1.87, 1.74, 1.94, 1.62, 2.06, 1.33, 1.96, 1.69, 1.70])
unpublished = np.array([2.39, 2.51, 2.86, 2.14, 2.56, 2.29, 2.49, 2.36, 2.58, 2.33, 2.62, 2.41, 1.94])

mean1, mean2 = np.mean(journals), np.mean(unpublished)
std1, std2 = np.std(journals, ddof=1), np.std(unpublished, ddof=1)
n1, n2 = len(journals), len(unpublished)

se = np.sqrt((std1**2 / n1) + (std2**2 / n2))
df = ((std1**2 / n1 + std2**2 / n2) ** 2) / (((std1**2 / n1) ** 2) / (n1 - 1) + ((std2**2 / n2) ** 2) / (n2 - 1))
t_critical = stats.t.ppf(1 - 0.05, df)

margin_of_error = t_critical * se
ci_lower, ci_upper = (mean1 - mean2) - margin_of_error, (mean1 - mean2) + margin_of_error

print(f"90% Confidence Interval for µ1 - µ2: ({ci_lower:.4f}, {ci_upper:.4f})")
```

```python
#6
import numpy as np
import scipy.stats as stats

method1 = np.array([14, 12, 18, 16, 15])
method2 = np.array([16, 15, 17, 16, 14])

differences = method1 - method2
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)
n = len(differences)

t_critical = stats.t.ppf(1 - 0.05/2, df=n-1)
margin_of_error = t_critical * (std_diff / np.sqrt(n))

ci_lower = mean_diff - margin_of_error
ci_upper = mean_diff + margin_of_error

print(f"95% Confidence Interval for the Mean Difference: ({ci_lower:.4f}, {ci_upper:.4f})")
```

```python
#7
import numpy as np
from scipy import stats

def calculate_std_confidence_interval(n: int, sample_std: float, confidence_level: float = 0.95) -> tuple:
    df = n - 1
    alpha = 1 - confidence_level
    chi2_lower = stats.chi2.ppf(alpha / 2, df)
    chi2_upper = stats.chi2.ppf(1 - alpha / 2, df)
    sample_var = sample_std ** 2
    var_lower = (n - 1) * sample_var / chi2_upper
    var_upper = (n - 1) * sample_var / chi2_lower
    std_lower = np.sqrt(var_lower)
    std_upper = np.sqrt(var_upper)
    return std_lower, std_upper

n = 100
sample_std = 0.01
confidence_level = 0.95

ci_lower, ci_upper = calculate_std_confidence_interval(n, sample_std, confidence_level)

print("Windshield Thickness Analysis")
print("-" * 30)
print(f"Sample Size: {n}")
print(f"Sample Standard Deviation: {sample_std} inches")
print(f"\n{confidence_level*100}% Confidence Interval for Population Standard Deviation:")
print(f"Lower bound: {ci_lower} inches")
print(f"Upper bound: {ci_upper} inches")



```
